```python
import findspark
import pandas as pd
import pyspark.sql.functions as f
from pyspark.sql import SparkSession

findspark.init()
findspark.find()
```

```python
spark = SparkSession.builder.appName("lab_Py_spark").master("local").getOrCreate()
```

```python
aircrafts = spark.read.format("csv").option("header", "true").option("inferSchema", "true").option("sep", ";").load("aircraft.csv")
```

```python
aircrafts.select("Вид воздушного судна").distinct().show()
```

```python
reggd_in_ru = aircrafts.filter(f.col("Вид воздушного судна") == "самолет").select(f.trim("Тип (наименование) воздушного судна")).distinct()
```

```python
# Построить список уникальных типов самолетов, зарегистрированных в России

reggd_in_ru.show(reggd_in_ru.count(), False)
```

```python
register_dates = aircrafts.select(f.to_date(f.col("дата действующего свидетельства о регистрации"), "dd.MM.yyyy").alias("date"))
```

```python
register_dates.select(f.max(f.col("date"))).show()
```

```python
#Какой тип самолета имеет самую позднюю дату выдчи сертификата

max_date = register_dates.collect()[0]
aircrafts.filter(f.to_date(f.col("дата действующего свидетельства о регистрации"), "dd.MM.yyyy") == max_date.date).select("Тип (наименование) воздушного судна").show()
```
