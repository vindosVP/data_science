```python
file_name = f'{path}/aircraft.csv'
separator = ';'

aircraft = pd.read_csv(
                    str(file_name),
                    sep=separator,
                    on_bad_lines="skip",
                    engine='python',
                )
```

```python
aircraft = aircraft.rename(
    columns={
        'Вид воздушного судна':'aircraft_type', 
        'Тип (наименование) воздушного судна':'aircraft_name', 
        'Опознавательные знаки':'signs', 
        'Серийный №':'serial_number', 
        'Идентификац. №':'identity_number', 
        'номер свидетельства о регистрации':'reg_number', 
        'дата действующего свидетельства о регистрации':'reg_date'
        }
)
```

```python
aircraft.info()
```

```python
aircraft['identity_number'] = aircraft['identity_number'].fillna('not_regd_in_RU')
aircraft.info()
```

```python
#Построить список уникальных типов самолетов, зарегистрированных в России

aircraft_rus = aircraft[aircraft['identity_number'] != 'not_regd_in_RU']
list(aircraft_rus[aircraft_rus['aircraft_type'] == 'самолет']['aircraft_name'].unique())
```

```python
#Какой тип самолета имеет самую позднюю дату выдчи сертификата
aircraft['reg_date'] = pd.to_datetime(aircraft['reg_date'], infer_datetime_format=True)
aircraft_date_sorted = aircraft.sort_values(by=['reg_date'], ascending=False)
aircraft_date_sorted['aircraft_type'][0]
```

```python
file_name = f'{path}/airports.csv'
separator = ','

airports = pd.read_csv(
                    str(file_name),
                    sep=separator,
                    on_bad_lines="skip",
                    engine='python',
                )
airports
```

```python
airports = airports.rename(
    columns={
        'Наименование аэропорта РФ':'airport_name', 
        'Свидетельство':'sertificate', 
        'Юрлицо':'owner', 
        'Тип':'type', 
        }
)
list(airports['airport_name'].unique())
```

```python
file_name = f'{path}/cargo transportation.csv'
separator = ';'

cargo_transportation = pd.read_csv(
                    str(file_name),
                    sep=separator,
                    on_bad_lines="skip",
                    engine='python',
                )

cargo_transportation
```

```python
cargo_transportation = cargo_transportation.rename(
    columns={
        'Наименование аэропорта РФ':'airport_name', 
        'Год периода данных':'year', 
        'Январь - Декабрь':'full_year_data', 
        }
)
cargo_transportation
```

```python
names_list = ['Январь', 'Февраль', 'Март', 'Апрель', 'Май', 'Июнь', 'Июль', 'Август', 'Сентябрь', 'Октябрь', 'Ноябрь', 'Декабрь', 'full_year_data']

cargo_transportation =  cargo_transportation.fillna(0.0)
for name in names_list:
    cargo_transportation[name] = cargo_transportation[name].apply(lambda x: str(x).replace(',', '.'))
    cargo_transportation[name] = cargo_transportation[name].apply(lambda x: str(x).replace(' ', ''))
    cargo_transportation[name] = cargo_transportation[name].apply(lambda x: 0 if x == '***' else x)
    cargo_transportation[name] = cargo_transportation[name].astype(float)

cargo_transportation.info()
```

```python
cargo_transportation['full_year_data'] = cargo_transportation.iloc[:, -13:-1].sum(axis=1)
cargo_transportation
```

```python
cargos_2018 = cargo_transportation[cargo_transportation['year'] == 2018]
cargos_2018 = cargos_2018[['airport_name', 'full_year_data']]
cargos_2018 = cargos_2018.rename(columns={'full_year_data':'count_cargos_2018'})
cargos_2018
```

```python

file_name = f'{path}/passenger transportation.csv'
separator = ';'

passenger_transportation = pd.read_csv(
                    str(file_name),
                    sep=separator,
                    on_bad_lines="skip",
                    engine='python',
                )

passenger_transportation
```

```python
passenger_transportation = passenger_transportation.rename(
    columns={
        'Наименование аэропорта РФ':'airport_name', 
        'Год периода данных':'year', 
        'Январь - Декабрь':'full_year_data' 
        }
)
passenger_transportation
```

```python
names_list = ['Январь', 'Февраль', 'Март', 'Апрель', 'Май', 'Июнь', 'Июль', 'Август', 'Сентябрь', 'Октябрь', 'Ноябрь', 'Декабрь', 'full_year_data']

cargo_transportation =  cargo_transportation.fillna(0.0)
for name in names_list:
    passenger_transportation[name] = passenger_transportation[name].apply(lambda x: str(x).replace(',', '.'))
    passenger_transportation[name] = passenger_transportation[name].apply(lambda x: str(x).replace(' ', ''))
    passenger_transportation[name] = passenger_transportation[name].apply(lambda x: 0 if x == '***' else x)
    passenger_transportation[name] = passenger_transportation[name].astype(float)

passenger_transportation.info()
```

```python
passenger_transportation['full_year_data'] = passenger_transportation.iloc[:, -13:-1].sum(axis=1)
passenger_transportation
```

```python
passangers_2018 = passenger_transportation[passenger_transportation['year'] == 2018]
passangers_2018 = passangers_2018[['airport_name', 'full_year_data']]
passangers_2018 = passangers_2018.rename(columns={'full_year_data':'count_passangers_2018'})
list(passangers_2018['airport_name'].unique())
```

```python
cargo_2018 = cargo_transportation[passenger_transportation['year'] == 2018]
cargo_2018 = cargo_2018[['airport_name', 'full_year_data']]
cargo_2018 = cargo_2018.rename(columns={'full_year_data':'count_cargo_2018'})
list(cargo_2018['airport_name'].unique())
```

```python
#грузы + пассажиры по аэропортам за 2018
temp_tab = passangers_2018
pass_cargo_2018 = temp_tab.merge(cargo_2018, on='airport_name', how='left')
pass_cargo_2018
```

```python
pass_cargo_2018['count_passangers_2018'] = pass_cargo_2018['count_passangers_2018'].astype(int)
pass_cargo_2018['count_passangers_2018'] = pass_cargo_2018['count_passangers_2018'].apply(lambda x: pd.NA if x == 0 else x)
pass_cargo_2018['count_cargo_2018'] = pass_cargo_2018['count_cargo_2018'].apply(lambda x: pd.NA if x == 0 else x)
```

```python
pass_cargo_2018
```

```python
#Перечислить аэропорты, где пассажиропоток больше медианы, а грузопоток меньше медианы
median_pass = pass_cargo_2018['count_passangers_2018'].sort_values().median()
median_cargo = pass_cargo_2018['count_cargo_2018'].sort_values().median()
print(median_pass, median_cargo)
list(pass_cargo_2018[pass_cargo_2018['count_passangers_2018'] > median_pass][pass_cargo_2018['count_cargo_2018'] < median_cargo]['airport_name'])
```
